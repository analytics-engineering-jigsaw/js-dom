# Event Listeners Lab

### Introduction

In this lesson, we'll practice adding event listeners to our HTML page.

### Getting Started

Let's begin by creating a new javascript file called `index.js` and then linking that file to the `post-page.html` file.

From there, let's get started with the following.  First add an event listener that logs `mountain picture clicked`, whenever someone clicks on the mountain picture.
